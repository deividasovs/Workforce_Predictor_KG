```python
# Create boxplots of each numerical column
df.plot(kind='box', column="subtotal")
df.plot(kind='box', column="transaction_count")
plt.show()
```

```python
# Remove outliers
df.loc[df['subtotal'] >= 11000, ['transaction_count', 'subtotal']] = np.nan
df.interpolate()
```

```python
df.plot(kind='box', column="subtotal")
df.plot(kind='box', column="transaction_count")
df.plot(kind='scatter', x='transaction_count', y='subtotal')

plt.show()
```
