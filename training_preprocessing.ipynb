# Load and pre-process our training dataset

```python
from sklearn import preprocessing
import numpy as np
import pandas as pd
```

```python
# Load from s3 instead?
df = pd.read_csv('.././Data/xMonths_SimulatedStaff.csv')
#../
```

```python

# No missing dates so we're just setting the time_idx equal to index. Should probably add some mathematical func here
# used to determine the sequence of samples.
# If no missing observations, the time index should increase by +1 for each subsequent sample. 
df["time_idx"] = df.index
```

```python
dateTime = pd.to_datetime(df['Timestamp']).dt
```

```python
# Adding some random constant key group
df["constant_group"] = "group_1"

#Change column value to all be floats
df['transaction_count'] = df['transaction_count'].astype(float)
```

```python
# Add year, month, day, and hour columns
df["year"] = dateTime.year
df["month"] = dateTime.month
df["day"] = dateTime.day
df["hour"] = dateTime.hour

df.head(10)
```

```python
# convert column workforce_type_1 to float
df['workforce_type_1'] = df['workforce_type_1'].astype(float)
df['workforce_type_2'] = df['workforce_type_2'].astype(float)
df['workforce_type_3'] = df['workforce_type_3'].astype(float)
df['workforce_type_4'] = df['workforce_type_4'].astype(float)
```

```python
df = df.drop(['Timestamp'], axis=1)
```

```python
# convert constant_group to numerical value
le = preprocessing.LabelEncoder()
df['constant_group'] = le.fit_transform(df['constant_group'])
df['holiday'] = le.fit_transform(df['holiday'])
```
